```python
# Phase 3: Data Integration & Feature Engineering

#The purpose of this phase is to combine the match-level and ball-by-ball datasets into a single analytical dataset and create additional features that can provide deeper insights into player and team performance.
```

```python
# Merge Datasets
ipl_df = deliveries.merge(
    matches,
    left_on="match_id",
    right_on="id",
    how="left"
)
```

```python
print("Merged Dataset Shape:")
print(ipl_df.shape)
```

```python
ipl_df.head()
```

```python
# ### Dataset Integration

# The deliveries dataset was merged with the matches dataset using the match identifier.

# This allows every ball-by-ball record to inherit match-level information such as:

# - Season
# - Venue
# - Winner
# - Toss Decision
# - Player of Match

# The resulting dataset provides a complete analytical view of IPL matches.
```

```python
# Assignment Requirement
# total_runs = batsman_runs + extra_runs
ipl_df["total_runs_scored"] = (
    ipl_df["batsman_runs"] +
    ipl_df["extra_runs"]
)
```

```python
ipl_df[
    [
        "batsman_runs",
        "extra_runs",
        "total_runs_scored"
    ]
].head()
```

```python
# ### Total Runs Feature

# A new variable called total_runs_scored was created.

# It represents the total runs obtained from a delivery including both batsman runs and extras.

# This feature is useful for run-scoring analysis.
```

```python
# Boundary Classification
ipl_df["is_boundary"] = np.where(
    ipl_df["batsman_runs"] >= 4,
    "Boundary",
    "Non-Boundary"
)
```

```python
ipl_df["is_boundary"].value_counts()
```

```python
# Percentage
round(
    ipl_df["is_boundary"]
    .value_counts(normalize=True) * 100,
    2
)
```

```python
# Over Phase
conditions = [
    ipl_df["over"].between(0, 5),
    ipl_df["over"].between(6, 14),
    ipl_df["over"].between(15, 19)
]

choices = [
    "Powerplay",
    "Middle Overs",
    "Death Overs"
]

ipl_df["over_phase"] = np.select(
    conditions,
    choices,
    default="Other"
)
```

```python
ipl_df["over_phase"].value_counts()
```

```python
round(
    ipl_df["over_phase"]
    .value_counts(normalize=True) * 100,
    2
)
```

```python
# ### Over Phase Classification

# Overs were divided into three phases:

# Powerplay:
# Overs 1–6

# Middle Overs:
# Overs 7–15

# Death Overs:
# Overs 16–20

# This classification helps analyze team performance across different stages of an innings.
```

```python
# Date Conversion
ipl_df["date"] = pd.to_datetime(
    ipl_df["date"]
)
```

```python
ipl_df["date"].dtype
```

```python
# ### Date Conversion

# The date column was converted into datetime format to enable time-based analysis.
```

```python
# Year Extraction
ipl_df["year"] = (
    ipl_df["date"]
    .dt.year
)
```

```python
ipl_df["year"].unique()
```

```python
# Month Extraction
ipl_df["month"] = (
    ipl_df["date"]
    .dt.month
)
```

```python
ipl_df["month"].value_counts()
```

```python
# ### Month Feature

# The month component was extracted from the match date.

# This can help identify monthly scoring trends and seasonality patterns.
```
