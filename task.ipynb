```python
# Phase 3: Data Integration & Feature Engineering

#The purpose of this phase is to combine the match-level and ball-by-ball datasets into a single analytical dataset and create additional features that can provide deeper insights into player and team performance.
```

```python
# Merge Datasets
ipl_df = deliveries.merge(
    matches,
    left_on="match_id",
    right_on="id",
    how="left"
)
```

```python
print("Merged Dataset Shape:")
print(ipl_df.shape)
```

```python
ipl_df.head()
```

```python
# ### Dataset Integration

# The deliveries dataset was merged with the matches dataset using the match identifier.

# This allows every ball-by-ball record to inherit match-level information such as:

# - Season
# - Venue
# - Winner
# - Toss Decision
# - Player of Match

# The resulting dataset provides a complete analytical view of IPL matches.
```

```python
# Assignment Requirement
# total_runs = batsman_runs + extra_runs
ipl_df["total_runs_scored"] = (
    ipl_df["batsman_runs"] +
    ipl_df["extra_runs"]
)
```

```python
ipl_df[
    [
        "batsman_runs",
        "extra_runs",
        "total_runs_scored"
    ]
].head()
```

```python
# ### Total Runs Feature

# A new variable called total_runs_scored was created.

# It represents the total runs obtained from a delivery including both batsman runs and extras.

# This feature is useful for run-scoring analysis.
```

```python
# Boundary Classification
ipl_df["is_boundary"] = np.where(
    ipl_df["batsman_runs"] >= 4,
    "Boundary",
    "Non-Boundary"
)
```

```python
ipl_df["is_boundary"].value_counts()
```

```python
# Percentage
round(
    ipl_df["is_boundary"]
    .value_counts(normalize=True) * 100,
    2
)
```

```python
# Over Phase
conditions = [
    ipl_df["over"].between(0, 5),
    ipl_df["over"].between(6, 14),
    ipl_df["over"].between(15, 19)
]

choices = [
    "Powerplay",
    "Middle Overs",
    "Death Overs"
]

ipl_df["over_phase"] = np.select(
    conditions,
    choices,
    default="Other"
)
```

```python
ipl_df["over_phase"].value_counts()
```

```python
round(
    ipl_df["over_phase"]
    .value_counts(normalize=True) * 100,
    2
)
```

```python
# ### Over Phase Classification

# Overs were divided into three phases:

# Powerplay:
# Overs 1–6

# Middle Overs:
# Overs 7–15

# Death Overs:
# Overs 16–20

# This classification helps analyze team performance across different stages of an innings.
```

```python
# Top 5 Batsmen
top_batsmen = (
    ipl_df.groupby("batter")["batsman_runs"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

top_batsmen
```

```python
top_batsmen_df = (
    ipl_df.groupby("batter")["batsman_runs"]
    .sum()
    .reset_index()
    .sort_values(
        by="batsman_runs",
        ascending=False
    )
    .head(5)
)

top_batsmen_df
```

```python
# Economy Rate Analysis
balls_bowled = (
    ipl_df.groupby("bowler")
    .size()
)
```

```python
runs_conceded = (
    ipl_df.groupby("bowler")
    ["total_runs_scored"]
    .sum()
)
```

```python
bowler_stats = pd.DataFrame({
    "balls_bowled": balls_bowled,
    "runs_conceded": runs_conceded
})
```

```python
bowler_stats = bowler_stats[
    bowler_stats["balls_bowled"] >= 500
]
```

```python
bowler_stats["overs_bowled"] = (
    bowler_stats["balls_bowled"] / 6
)
```

```python
bowler_stats["economy_rate"] = np.round(
    bowler_stats["runs_conceded"] /
    bowler_stats["overs_bowled"],
    2
)
```

```python
# Top Economical Bowlers
economy_df = (
    bowler_stats
    .sort_values(
        by="economy_rate"
    )
)

economy_df.head(10)
```

```python
# ### Economy Rate Analysis

# Economy rate measures the average number of runs conceded per over.

# Only bowlers with at least 500 deliveries were considered to ensure a fair comparison.
```

```python
# Highest Win Percentage Per Season
```

```python
# Matches Played
team1 = matches[
    ["season","team1"]
].rename(
    columns={
        "team1":"team"
    }
)

team2 = matches[
    ["season","team2"]
].rename(
    columns={
        "team2":"team"
    }
)

all_matches = pd.concat(
    [team1, team2]
)
```

```python
# Matches Played Count
matches_played = (
    all_matches
    .groupby(
        ["season","team"]
    )
    .size()
    .reset_index(
        name="matches_played"
    )
)
```

```python
# Wins Count
wins = (
    matches.groupby(
        ["season","winner"]
    )
    .size()
    .reset_index(
        name="wins"
    )
)

wins.rename(
    columns={
        "winner":"team"
    },
    inplace=True
)
```

```python
# Merge
season_stats = (
    matches_played.merge(
        wins,
        on=[
            "season",
            "team"
        ],
        how="left"
    )
)
```

```python
# Fill Missing Wins
season_stats["wins"] = (
    season_stats["wins"]
    .fillna(0)
)
```

```python
# Win Percentage
season_stats["win_percentage"] = (
    season_stats["wins"] /
    season_stats["matches_played"]
) * 100
```

```python
# Highest Win % Team Each Season
best_team_each_season = (
    season_stats
    .loc[
        season_stats.groupby(
            "season"
        )[
            "win_percentage"
        ].idxmax()
    ]
)
```

```python
# Final Result
best_team_each_season[
    [
        "season",
        "team",
        "matches_played",
        "wins",
        "win_percentage"
    ]
]
```

```python
# ### Seasonal Win Percentage Analysis

# For each IPL season, the team with the highest win percentage was identified.

# Win percentage was calculated as:

# Win Percentage = (Wins / Matches Played) × 100
```

```python
# Graph 1: Toss Impact on Match Outcome
```

```python
toss_analysis = (
    matches.groupby("toss_decision")
    .apply(
        lambda x:
        (
            x["toss_winner"] ==
            x["winner"]
        ).mean()*100
    )
    .reset_index(
        name="win_percentage"
    )
)

toss_analysis
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=toss_analysis,
    x="toss_decision",
    y="win_percentage"
)

plt.title(
    "Win Percentage by Toss Decision"
)

plt.xlabel(
    "Toss Decision"
)

plt.ylabel(
    "Winning Percentage"
)

plt.show()
```

```python
# Teams choosing to field first appear to have a higher winning percentage than teams choosing to bat first.

# This suggests that chasing a target may provide a strategic advantage in IPL matches.
```

```python
plt.figure(figsize=(10,6))

top10.sort_values().plot(
    kind="barh"
)

plt.title(
    "Top 10 Run Scorers in IPL"
)

plt.xlabel("Runs")
plt.ylabel("Batter")

plt.show()
```

```python
# Graph 4: Seasonal Batting Trends
season_runs = (
    ipl_df.groupby("season")
    ["total_runs_scored"]
    .sum()
)
```

```python
season_matches = (
    matches.groupby("season")
    .size()
)
```

```python
avg_runs = (
    season_runs /
    season_matches
)
```

```python
plt.figure(figsize=(12,6))

avg_runs.plot(
    marker="o"
)

plt.title(
    "Average Runs Per Match by Season"
)

plt.xlabel("Season")
plt.ylabel("Average Runs")

plt.grid(True)

plt.show()
```

```python
# Average runs per match have generally increased over time.

# Modern IPL seasons are more batting friendly than earlier editions.
```

```python
# Graph 5: Regression Trend
trend_df = avg_runs.reset_index()

trend_df.columns = [
    "season",
    "avg_runs"
]
```

```python
plt.figure(figsize=(10,6))

sns.regplot(
    data=trend_df,
    x=trend_df.index,
    y="avg_runs"
)

plt.title(
    "IPL Scoring Trend"
)

plt.xlabel("Season")
plt.ylabel("Average Runs")

plt.show()
```

```python
# The positive regression slope suggests that IPL scoring rates are increasing over time.

# Teams are adopting more aggressive batting strategies.
```

```python
# Graph 7: Death Over Performance
death_df = ipl_df[
    ipl_df["over_phase"]
    == "Death Overs"
]
```

```python
top_teams = (
    death_df.groupby(
        "batting_team"
    )
    ["total_runs_scored"]
    .sum()
    .sort_values(
        ascending=False
    )
    .head(6)
    .index
)
```

```python
death_top = death_df[
    death_df[
        "batting_team"
    ]
    .isin(top_teams)
]
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    data=death_top,
    x="batting_team",
    y="total_runs_scored"
)

plt.title(
    "Death Over Performance"
)

plt.xticks(rotation=45)

plt.show()
```

```python
# Teams with higher median values score more consistently during death overs.

# Teams with wider boxplots exhibit greater variability and explosive scoring potential.
```
