```python
# Phase 2: Data Cleaning & Preprocessing
```

```python
print("Matches Dataset Missing Values")
matches.isnull().sum().sort_values(ascending=False)
```

```python
matches["city"].isnull().sum()
```

```python
matches["city"] = matches["city"].fillna("Unknown")
```

```python
matches["player_of_match"].isnull().sum()
```

```python
matches["player_of_match"] = matches["player_of_match"].fillna("No Award")
```

```python
matches[matches["winner"].isnull()]
```

```python
matches["winner"] = matches["winner"].fillna("No Result")
```

```python
matches[
    ["target_runs","target_overs"]
].isnull().sum()
```

```python
matches["target_runs"] = matches["target_runs"].fillna(0)
matches["target_overs"] = matches["target_overs"].fillna(0)
```

```python
matches["method"].value_counts(dropna=False)
```

```python
deliveries["dismissal_kind"] = (
    deliveries["dismissal_kind"]
    .fillna("not out")
)
```
